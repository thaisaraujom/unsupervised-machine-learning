Se temos uma variável aleatória ($U$) que é uniformemente distribuída entre 0 e 1, a média ($E[U]$) é dada pela integral:

$ E[U] = \int_{0}^{1} u f(u) du $

Onde ($f(u)$) é a função densidade de probabilidade da distribuição uniforme, que é 1 para ($u$) entre 0 e 1.

A função densidade de probabilidade ($ f(u) $) para uma distribuição uniforme entre 0 e 1 é constante e igual a 1 para todo ($ u $) no intervalo [0,1]. Fora desse intervalo, ($ f(u) $) é 0.

Então, a integral que queremos resolver é:

$ E[U] = \int_{0}^{1} u \times 1 du $

Isso é equivalente a:

$ E[U] = \int_{0}^{1} u du $

Agora, integrando $( u )$ em relação a $( u )$ no intervalo \[0,1], tem-se:


$E[U] = \int_{0}^{1} u du $

Ao integrar $( u )$ em relação a $( u )$, obtemos:

$ \frac{u^2}{2} \Big|_0^1 $

Agora, substituindo os limites de integração:

$ \frac{1^2}{2} - \frac{0^2}{2} = \frac{1}{2} $

Portanto, a média $( E[U] )$ de uma variável aleatória $( U )$ uniformemente distribuída entre 0 e 1 é $( \frac{1}{2} )$.

A variância $( \sigma^2 )$ de uma variável aleatória $( U )$ é definida como:

$ \sigma^2[U] = E[U^2] - (E[U])^2 $

Onde $( E[U^2] )$ é o segundo momento e $( E[U] )$ é a média que foi calculada.

Para encontrar $( E[U^2] )$, precisamos resolver a seguinte integral:

$ E[U^2] = \int_{0}^{1} u^2 du $

Resolvendo a integral, tem-se:

 $E[U^2] = \int_{0}^{1} u^2 du = \frac{u^3}{3} \Big|_0^1 = \frac{1^3}{3} - \frac{0^3}{3} = \frac{1}{3} $

Agora que temos $( E[U^2] = \frac{1}{3} )$ e $( E[U] = \frac{1}{2} )$, podemos calcular a variância $( \sigma^2[U] )$:

 $\sigma^2[U] = E[U^2] - (E[U])^2 = \frac{1}{3} - \left(\frac{1}{2}\right)^2 = \frac{1}{3} - \frac{1}{4} = \frac{1}{12} $

Portanto, a variância $( \sigma^2[U] )$ de uma variável aleatória $( U )$ uniformemente distribuída entre 0 e 1 é $( \frac{1}{12} )$.

Agora, considerando a soma de $( n )$ variáveis aleatórias uniformes independentes, temos que:

Se $( X )$ é a soma de $( n )$ variáveis aleatórias uniformes independentes $( U_1, U_2, \dots, U_n )$, então:

$ E[X] = E[U_1 + U_2 + \dots + U_n] $

Usando a linearidade da expectativa:

$ E[X] = E[U_1] + E[U_2] + \dots + E[U_n] $

Como todas as $( U_i )$ são idênticas e distribuídas uniformemente entre 0 e 1, temos:

$ E[X] = n \times E[U] = n \times \frac{1}{2} = \frac{n}{2} $

Para a variância:

$ \sigma^2[X] = Var(U_1 + U_2 + \dots + U_n) $

E como as variáveis aleatórias são independentes:

$ \sigma^2[X] = Var(U_1) + Var(U_2) + \dots + Var(U_n) $

Novamente, como todas as $( U_i )$ são idênticas:

$ \sigma^2[X] = n \times Var(U) = n \times \frac{1}{12} = \frac{n}{12} $


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

size = 10000
n_values = range(1, 13)

# Definindo o tamanho da figura
plt.figure(figsize=(20, 15))

for i, n in enumerate(n_values, start=1):
    # Gere n conjuntos de amostras uniformes e some-os
    samples = sum([stats.uniform.rvs(size=size, random_state=None) for _ in range(n)])

    # Cálculo da média e variância
    mean = samples.mean()
    variance = samples.var()

    # Cálculo do segundo momento
    second_moment = np.mean(samples**2)

    print(f"For n={n}:")
    print(f"Mean E[X] = {mean}")
    print(f"Variance σ^2[X] = {variance}")
    print(f"Second Moment E[X^2] = {second_moment}")
    print("-" * 40)

    # Definindo o subplot para PDF
    plt.subplot(4, 6, 2*i-1)  # 4 linhas e 6 colunas

    # Plot histograma (PDF)
    plt.hist(samples, bins=int(1+3.22*np.log(size)), density=True, alpha=0.6, label=f'n={n}')

    # Plot PDF da distribuição normal para comparação
    mu = stats.norm.mean(loc=n/2, scale=np.sqrt(n/12))
    sigma = stats.norm.std(loc=n/2, scale=np.sqrt(n/12))

    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    plt.plot(x, stats.norm.pdf(x, loc=mu, scale=sigma), 'r-', lw=2)

    plt.title(f'PDF for n = {n}')
    plt.legend()

    # Definindo o subplot para CDF
    plt.subplot(4, 6, 2*i)  # 4 linhas e 6 colunas

    # Plot CDF da distribuição normal para comparação
    plt.hist(samples, bins=int(1+3.22*np.log(size)), density=True, alpha=0.6, label=f'n={n}', cumulative=True, histtype='step')
    plt.plot(x, stats.norm.cdf(x, loc=mu, scale=sigma), 'r-', lw=2)

    plt.title(f'CDF for n = {n}')
    plt.legend()

# Ajustando o layout para evitar sobreposição
plt.tight_layout()
plt.show()
```

1. **Média, Variância e Segundo Momento de uma v.a. uniformemente distribuída entre 0 e 1**:
   - A média $(E[X])$ de uma variável aleatória $(X)$ uniformemente distribuída entre 0 e 1 é $(\frac{1}{2})$.
   - A variância $(\sigma^2[X])$ é $(\frac{1}{12})$.
   - O segundo momento $(E[X^2])$ é $(\frac{1}{3})$.

2. **Média e Variância da soma de $(n)$ v.a. uniformemente distribuídas**:
   - A média da soma é $(n \times \frac{1}{2} = \frac{n}{2})$.
   - A variância da soma é $(n \times \frac{1}{12} = \frac{n}{12})$.

3. **Gráficos das funções densidade de probabilidade $(f_X(X))$ e distribuição de probabilidade $(F_X(X))$ para $(n = 1,2, \ldots, 12)$**:
   - À medida que $(n)$ aumenta, a forma da distribuição da soma das variáveis aleatórias tende a se aproximar de uma distribuição normal devido ao Teorema do Limite Central. Isso significa que, mesmo que as variáveis originais não sejam normalmente distribuídas, a distribuição da sua soma tende a ser normal à medida que o número de variáveis aumenta.

4. **Para qual distribuição a v.a. $(X)$ tende quando o valor de $(n)$ aumenta?**:
   - A variável aleatória $(X)$ tende a uma distribuição normal à medida que $(n)$ aumenta. Isso é uma consequência do Teorema do Limite Central.

5. **Por que isto ocorre?**:
   - Isso ocorre devido ao Teorema do Limite Central, que afirma que a distribuição da soma (ou média) de um grande número de variáveis aleatórias independentes e identicamente distribuídas, cada uma com média e variância finitas, será aproximadamente normal, independentemente da forma da distribuição original das variáveis aleatórias.

6. **Há uma expressão para a média e a variância em função de $(n)$?**:
   - Sim, como mencionado anteriormente:
     - Média: $(E[X] = \frac{n}{2})$
     - Variância: $(\sigma^2[X] = \frac{n}{12})$
