Seja um v.a. $X$ com Função Densidade de Probabilidade (FDP) Normal com média zero e variância igual a 1. Gere uma sequência desta distribuição com 10.000 pontos.

a. Obtenha histogramas para os 10 primeiros pontos, os 100 primeiros pontos, os 1.000 primeiros pontos e os 10.000 pontos. Analise os resultados.
   
b. Calcule a média e a variância considerando-se: os 10 primeiros pontos, os 100 primeiros pontos, os 1.000 primeiros pontos, e os 10.000 pontos. Analise os resultados. Sugestão: plote os gráficos da média e variância em função do número de pontos utilizados.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Geração de uma sequência da distribuição normal com 10.000 pontos
x = stats.norm.rvs(loc=0, scale=1, size=10000, random_state=None)

# Definindo os tamanhos das sequências
sizes = [10, 100, 1000, 10000]
```

```python
# a)
# Plotando os histogramas
plt.figure(figsize=(15, 10))

for i, size in enumerate(sizes, start=1):
    plt.subplot(2, 2, i)
    plt.hist(x[:size], bins=int(1+3.22*np.log(size)), density=True, alpha=0.6)
    plt.title(f'Histogram for first {size} points')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Análise dos resultados:

- 10 primeiros pontos: Com tão poucos pontos, o histograma não se parece muito com uma distribuição normal. Há muita variabilidade e é difícil tirar conclusões.

- 100 primeiros pontos: Começamos a ver uma forma que se assemelha mais a uma distribuição normal, mas ainda há alguma variabilidade.

- 1.000 primeiros pontos: O histograma agora se parece muito mais com uma distribuição normal. A maioria dos pontos está concentrada em torno da média, com menos pontos nas caudas.

- 10.000 pontos: Com essa quantidade de pontos, o histograma se assemelha fortemente a uma distribuição normal. A forma é suave e a maioria dos pontos está concentrada em torno da média.

À medida que aumentamos o número de pontos, o histograma se aproxima cada vez mais da forma esperada de uma distribuição normal, o que é uma ilustração do Teorema do Limite Central em ação.

```python
# b)
# Plotando os histogramas
plt.figure(figsize=(15, 10))

# Listas para armazenar os valores de média e variância
means = []
variances = []

# Calculando a média e variância para cada tamanho de sequência
for size in sizes:
    means.append(np.mean(x[:size]))
    variances.append(np.var(x[:size]))

# Plotando os gráficos
plt.figure(figsize=(15, 5))

# Gráfico da média
plt.subplot(1, 2, 1)
plt.plot(sizes, means, marker='o', linestyle='-')
plt.title('Mean vs. Number of Points')
plt.xlabel('Number of Points')
plt.ylabel('Mean')
plt.grid(True)

# Gráfico da variância
plt.subplot(1, 2, 2)
plt.plot(sizes, variances, marker='o', linestyle='-')
plt.title('Variance vs. Number of Points')
plt.xlabel('Number of Points')
plt.ylabel('Variance')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# 1
print("Última média calculada:", means[-1])
print("Última variância calculada:", variances[-1])
```

```python
# 2
print("Última média calculada:", means[-1])
print("Última variância calculada:", variances[-1])
```

Análise dos resultados:

- À medida que o número de pontos aumenta, a média e a variância tendem a convergir para os valores teóricos da distribuição normal, que são 0 e 1, respectivamente.
- Com um número menor de pontos, há mais variabilidade nos valores calculados de média e variância. Isso é esperado, pois com menos dados, há mais chance de obter valores atípicos que podem afetar a média e a variância.
- À medida que o número de pontos aumenta, a variabilidade nos valores calculados diminui e eles se aproximam dos valores teóricos.

Os gráficos ilustram a ideia de que, à medida que temos mais dados, nossas estimativas se tornam mais precisas e confiáveis.
